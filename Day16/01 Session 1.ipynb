    1. 2x +  y =  3

```python
x = np.arange(-10, 11)
y = 3 - (2 * x)
print(x)
print(y)

plt.plot(x, y)
```

```python
x = np.arange(-10, 11)
y = (-2 - (5 * x))/3
print(x)
print(y)

plt.plot(x, y)
```

```python
# 5x + 3y = 2
x = np.arange(-10, 11)
y = (2 - (5 * x))/3
print(x)
print(y)

plt.plot(x, y)
```

```python
# y = mx + c
# c = 0; m=2;
x = np.arange(-10, 11)
y = (2 * x) + 0
print(x)
print(y)
plt.plot(x, y)
```

```python
# y = mx + c
# c = 2; m=2;
x = np.arange(0, 11)
y = (2 * x) + 2
print(x)
print(y)
plt.plot(x, y)
```
