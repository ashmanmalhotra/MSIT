# Filling NA Values

```python
# Generate an Array and DataFrame from the array of shape (8,8)
dummyval = np.random.randint(0, 30, 64).reshape(8,8)
dummyval
```

```python
df = pd.DataFrame(data=dummyval, columns=['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'])
df
```

```python
# Let's fill some cells in DataFrame with NaN values
# This loop will fill 12 cells with NaN value - Loop will iterate 12 times
for i in range(0, 13):
    row, column = np.random.randint(0,7), np.random.randint(0,7)
    df.iloc[row, column] = np.nan

df.astype = np.int
df
```

```python
# Let's Fill NA values
# fillna()
df.fillna(0)
```

```python
df
```

```python
df.fillna(method="ffill")
```

```python
df
```

```python
df.fillna(method="ffill", axis=1)
```

```python
df
```

```python
df.fillna(method="bfill")
```
