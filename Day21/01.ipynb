```python
X = df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]
y = df['Price']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
lm.intercept_
```

```python
lm.coef_
```

```python
cdf = pd.DataFrame(lm.coef_, X.columns, columns=['Coefficient'])
cdf
```

```python
predictions = lm.predict(X_test)
predictions
```

```python
y_test.head()
```

```python
plt.scatter(y_test, predictions)
```

```python
sns.distplot((np.array(tuple(y_test)) - predictions), bins=50);
```

```python
from sklearn import metrics
```

```python
"MAE : ", metrics.mean_absolute_error(y_test, predictions)
```

```python
"MSE : ", metrics.mean_squared_error(y_test, predictions)

81349.24092

```

```python
"RMSE : ", (metrics.mean_squared_error(y_test, predictions)) ** (1/2)
```

```python
"RMSE : ", np.sqrt(metrics.mean_squared_error(y_test, predictions))
```

```python
resdf = pd.DataFrame({"YTest": np.array(y_test), "Predictions":predictions})
```
